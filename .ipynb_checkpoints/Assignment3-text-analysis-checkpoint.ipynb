```python
##### Changing classification from string to numbers##########


# data = df_train.gender.replace(to_replace= {'f':0, 'm':1}, inplace=False)
# data

# def alpha_to_num(df):
#     #(m,f) -> (1,0)
#     lines = df['gender'].values.tolist()
#     for line in range(len(lines)): 
#         if lines[line] == 'm':
#             lines[line] = 1
#         else: 
#             lines[line] = 0
#     return lines

# for line in range(len(gender)):
#     print(gender[line], df_train.iloc[line][1])
# len(gender)

y_gender = df_train["gender"].factorize()[0]
# y_gender
y_gender.shape

```

```python
def clean_data(df):
    # Initialize an empty list to store cleaned stories.
    all_stories = list()
    
    # Extract stories from the dataframe and convert them into a list.
    lines = df["story"].values.tolist()

    # Loop through each story in the list.
    for text in lines:
        # Remove punctuation from the story.
        text = re.sub(r"[,.\"!@#$%^&*(){}?/;`~:<>+=-]", "", text)
        
        # Tokenize the cleaned text.
        tokens = list(ht.tokenize(text))
        
        # Create a translation table to remove signs, numbers, and other unwanted characters.
        transTable = str.maketrans('', '', r"!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~|[0-9]")
        
        # Use the translation table to clean each token.
        cleaned = [w[1].translate(transTable) for w in tokens]
        
        # Remove non-Hebrew words (assuming words that are ASCII are not Hebrew).
        words = [word for word in cleaned if not word.isascii()]
        
        # Convert the list of cleaned words back into a single string.
        words = ' '.join(words)
        
        # Append the cleaned story to the final list.
        all_stories.append(words)
    
    # Return the list of cleaned stories.
    return all_stories

# Call the clean_data function on the training dataframe to get cleaned stories.
X_all_stories = clean_data(df_train)

# Uncomment the following lines to inspect the results.
# X_all_stories[:20]
# len(all_stories)
# type(all_stories)
```

```python
##### Combine the 2 ########
d = {'story': X_all_stories, 'gender': y_gender}
train_data = pd.DataFrame(d)
train_data.head()
```

```python
# Initialize the TfidfVectorizer with specified parameters.
# max_features: Limit the number of features (words) to the top 1200 by frequency.
# decode_error: If there's a decoding error, replace the problematic character.
# sublinear_tf: Use sublinear scaling, i.e., replace tf with 1 + log(tf).
tfIdf = TfidfVectorizer(max_features= 1200, decode_error="replace", sublinear_tf=True)

# Fit the vectorizer on the 'story' column of the training data and transform the stories into vectors.
# The result is converted to an array for further processing.
X_train = tfIdf.fit_transform(train_data['story']).toarray()
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_gender, test_size=0.2, random_state=42, stratify=y_gender)
```

```python
# List of models to train
models = [
    DecisionTreeClassifier(criterion='entropy', random_state=41),
    LinearSVC(),
    GaussianNB(),
    MultinomialNB()
]

# Initialize an empty list to store F1 scores for each model
f1_scores = []

# Train and evaluate each model
for mod in models:
    # Train the model on the training split
    mod.fit(X_train_split, y_train_split)
    
    # Use cross-validation to evaluate the F1 score, making sure that n_splits is not greater than the minimum class count
    n_splits = min(5, np.min(np.bincount(y_gender)))  # Set n_splits to be no greater than the smallest class size
    cv_f1 = np.mean(cross_val_score(mod, X_train, y_gender, cv=n_splits, scoring='f1_macro'))
    
    # Predict the labels for the validation split
    y_pred = mod.predict(X_val_split)
    
    # Evaluate the model
    print(f"{mod}")
    print("Cross-Validation F1:", cv_f1)
    print("Validation Accuracy: ", accuracy_score(y_val_split, y_pred))
    print("Validation F1: ", f1_score(y_val_split, y_pred, average='macro'))
    print("Validation Precision: ", precision_score(y_val_split, y_pred, average='macro'))
    print()
    
    # Append the cross-validation F1 score to the f1_scores list
    f1_scores.append(cv_f1)

# Identify the model with the highest F1 score
best_model = models[np.argmax(f1_scores)]

# Output the best model
print("\nBest model based on Cross-Validation F1:", best_model)
```

```python

# define the parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'loss': ['hinge', 'squared_hinge']
}

# create a GridSearchCV object
grid_search = GridSearchCV(LinearSVC(), param_grid, cv=5, scoring='f1_macro')

# fit the model to the training split
grid_search.fit(X_train_split, y_train_split)
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Print the best hyperparameters
print("Best Parameters:", best_params)

# Perform cross-validation on the entire training set using the best model
cv_f1 = np.mean(cross_val_score(best_model, X_train, y_gender, cv=5, scoring='f1_macro'))
print("Cross-Validation F1 Score on Entire Training Set:", cv_f1)

# Use the best model to make predictions on the validation set
y_pred_val = best_model.predict(X_val_split)

# Evaluate the model on the validation set
print("Validation F1: ", f1_score(y_val_split, y_pred_val, average='macro'))
print("Validation Accuracy: ", accuracy_score(y_val_split, y_pred_val))
print("Validation Precision: ", precision_score(y_val_split, y_pred_val, average='macro'))
```

```python
# Fit best model on training data.
best_model.fit(X_train, y_gender)  

# Predict labels for training data using best model.
y_pred = best_model.predict(X_train)

# the macro F1 score of the best model.
print(f1_score(y_gender, y_pred, average='macro'))

print(pd.DataFrame({'y_train': y_gender, 'y_pred': y_pred}).head(10))  
print(pd.DataFrame({'y_train': y_gender, 'y_pred': y_pred}).tail(10))  

# the shape of the DataFrame, indicating the number of rows and columns.
print(pd.DataFrame({'y_train': y_gender, 'y_pred': y_pred}).shape)
```

```python
# Transform the 'story' column of the test dataset using the tfIdf vectorizer.
X_test_transformed = tfIdf.transform(df_test['story'])

y_pred_test = best_model.predict(X_test_transformed)

# Create a new column for predicted gender labels.
# Convert the numerical labels to their string counterparts ('m' for male and 'f' for female).
df_test['predicted_gender'] = pd.Series(y_pred_test).replace([0, 1], ['m', 'f'])

# the count of each predicted gender in the test DataFrame.
print("Predicted Gender Counts:")
print(df_test['predicted_gender'].value_counts())


print("\nFirst 5 Rows:")
print(df_test['predicted_gender'].head())

print("\nLast 5 Rows:")
print(df_test['predicted_gender'].tail())
```
